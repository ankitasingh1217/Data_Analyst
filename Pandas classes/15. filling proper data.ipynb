```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=['day']) 
# day column will be converted to date format which was previously str
df

```

```python
new_df=df.fillna({
    
    'temperature':0,
    'windspeed':0,
    'event':'no event'
       
})
new_df
```

# here after 1st date tempr. is zero so to manage this at all ...we now to carry forward previous value we use following method.

```python
new_df=df.fillna(method="ffill")
new_df
```

# look here 32 and next 32 in temprature... 28 and next 28 so on..

# now to carry farward next days value to previous

```python
new_df=df.fillna(method='bfill')
new_df
```

# here next values filled to previous

# now we will see axis='columns' option...this is used to copy data horizontally from next column ..

# look below in temperature 2 row 9 is copied from windspeed colum..that is horizontal copy

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

# limiting ffill:-

# if we ffill(farward fill) all NAN are filled by previous values.

# i want only one NAN should be filled by previous value

# this can be achieved by limit=1


```python
new_df=df.fillna(method="ffill")
new_df
```

# in aboove case value 32(inn temperture column) is filled from 4 row  to 5 & 6 row..so we can restrict to only one i.e. 5th row by limit=1

```python
new_df=df.fillna(method="ffill",limit=1)
new_df
```

# look in above output  6th row temprature is NAN as we limited 32 to 5th only.and even 3rd row  of windspped value 7 copied once.
