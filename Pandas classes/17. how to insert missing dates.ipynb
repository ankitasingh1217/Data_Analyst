```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=['day']) 
# day column will be converted to date format which was previously str
df.set_index('day',inplace=True)
df
```

# here we are inserting missing dates in our day column look for example  after 1st jan there is 4th jan so to fill between days we do following


```python
dt=pd.date_range('01-01-2017','01-11-2017')
idx=pd.DatetimeIndex(dt)
df=df.reindex(idx)
df
```
