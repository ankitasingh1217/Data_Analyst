# Now we will convert NAN values in data frame to meaningful values

# fillna() method

```python
import pandas as pd
df=pd.read_csv("weather_data.csv",parse_dates=['day']) 
# day column will be converted to date format which was previously str
df

# parse_dates is used to convert string to date format
```

```python
# here we are replacing all NAN values as 0 (Zero) . using fillna function 
#and storing it with new_df (data frame) 
# as i dont want to change orignal data frame

new_df=df.fillna(0)
new_df
```

# see NAN  is replaced with zero(0)

#  Now suppose you donot want zero(0) every where then we can do following step

# here we are passing dictionary to fillna () function

```python
new_df=df.fillna({
    
    'temperature':0,
    'windspeed':0,
    'event':'no event'
       
})
new_df
```

# here we have changed temprature and windespeed with "0" and event with "no event"
